```python
#아래는 Pandas를 사용하여 A 열로 그룹화하고 B, C, D 열에 따라 count하는 코드입니다.

# 예제 데이터프레임 생성

# A 컬럼으로 그룹화하고 B, C, D 컬럼에 대해 count
# extrc_word_nm : 둘레길 단어
# auser_sxdst_cd : 성별
# age_group_cd : 나이 그룹
# auser_ara_ctprv_nm : 거주지
# grouped = df.groupby('extrc_word_nm').agg({'auser_sxdst_cd': 'count', 'age_group_cd': 'count', 'auser_ara_ctprv_nm': 'count'}).reset_index()
# grouped.columns = ['extrc_word_nm', 'Count_sex', 'Count_age_group', 'Count_user_live']

# 필터링
# filtered_df = df[(df['age_group_cd'] == 2) | (df['age_group_cd'] == 3)]


df = pd.read_csv('/content/sample_data/FPL010101txt.txt', encoding ='cp949')

# 나이 그룹 별로 그룹화한 후에 count
grouped = df.groupby(['age_group_cd']).size().reset_index(name='count')

print(grouped)


df2 = pd.read_csv('/content/sample_data/FPL060101txt.txt', encoding ='utf-8')

# 나이 그룹 별로 그룹화한 후에 count
grouped2 = df2.groupby(['age_group_cd']).size().reset_index(name='count')

print(grouped2)



import matplotlib.pyplot as plt

# 파이차트를 위한 데이터 준비
labels = grouped['age_group_cd']
sizes = grouped['count']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgrey']
explode = (0.1, 0, 0, 0, 0)  # 파이차트에서 특정 조각을 튀어나오게 만듭니다. 여기서는 첫 번째 값을 튀어나오게 합니다.

# 파이차트 그리기
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # 파이차트가 원형이 되도록 설정합니다.

# 제목 추가
plt.title('Dul-re-gil Age Group Distribution')

# 그래프 보이기
plt.show()
```
